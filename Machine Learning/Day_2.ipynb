How does Machine Learning algorithm work?
In this objective, the lesson would likely introduce the fundamental steps involved in the machine learning process:

> a. Data Collection: Gathering relevant data is the first step. For instance, consider a dataset of housing prices with features like the size of the house, the number of bedrooms, and the location.

>b. Data Preprocessing: Cleaning and preparing the data for analysis is essential. This step involves handling missing values, scaling features, and encoding categorical variables.

>c. Model Training: Selecting an appropriate machine learning algorithm (e.g., linear regression) and fitting it to the training data. The algorithm learns the underlying patterns and relationships in the data.

>d. Model Evaluation: Assessing the performance of the trained model using evaluation metrics like mean squared error (MSE) or R-squared (coefficient of determination).

>e. Model Deployment: After successfully training and evaluating the model, it can be deployed to make predictions on new, unseen data.

Learning about Linear Regression as the simplest example of a Machine Learning model
Linear regression is a straightforward and widely-used machine learning model for regression tasks. It models the relationship between a dependent variable (target) and one or more independent variables (features) as a linear equation.

For a simple example, consider predicting house prices based on the size of the house. The linear regression model will try to find the best-fitting line that represents the relationship between the house's size (feature) and its price (target). The equation of a simple linear regression is:

y = mx + b

where y is the predicted price, x is the size of the house, m is the slope (weight) of the line, and b is the y-intercept.

ore ML concepts
This objective would introduce key concepts and terminologies used throughout machine learning:

>a. Features and Labels: Features are the input variables used to make predictions, while labels are the output or target variables that the model tries to predict. In the housing price example, the features are the size of the house, number of bedrooms, etc., and the label is the price.

>b. Training Data and Testing Data: The dataset is typically divided into training and testing sets. The model learns from the training data and is evaluated on the testing data to assess its generalization ability.

>c. Overfitting and Underfitting: Overfitting occurs when a model performs well on the training data but poorly on unseen data, indicating that it has memorized the training data rather than learned general patterns. Underfitting happens when a model is too simple to capture the underlying patterns in the data.

>d. Generalization: A model that performs well on unseen data is said to generalize well. Generalization is the ultimate goal in machine learning.

How Machine Learning Work!
This objective could provide a more detailed explanation of various learning paradigms and algorithms:

>a. Supervised Learning: In supervised learning, the model is trained on labeled data, where the correct answers (labels) are provided. Examples include regression and classification tasks.

>b. Unsupervised Learning: In unsupervised learning, the model is trained on unlabeled data, and it learns patterns and structures from the data without explicit target labels. Examples include clustering and dimensionality reduction.

>c. Reinforcement Learning: Reinforcement learning involves training a model to interact with an environment and learn from feedback in the form of rewards or penalties. It is often used in areas like robotics and game playing.



Model Evaluation/Validation
Model evaluation is crucial to ensure that the trained model performs well on unseen data. Techniques for model evaluation include:

>a. Accuracy: For classification tasks, accuracy measures the percentage of correctly predicted instances.

>b. Mean Squared Error (MSE): For regression tasks, MSE calculates the average squared difference between predicted and actual values.

>c. Cross-Validation: To avoid overfitting, cross-validation divides the dataset into multiple subsets, trains and tests the model on different combinations of these subsets, and averages the performance scores.

>d. Hold-Out Validation: The dataset is split into a training set and a testing set. The model is trained on the training set and evaluated on the testing set.

>b. Grid Search: A method that systematically searches through a predefined set of hyperparameter combinations to find the best-performing model.

Multiple Linear regression
This objective extends the concept of linear regression to include multiple independent variables.

For instance, consider predicting house prices based on both the size and the number of bedrooms. The multiple linear regression model will fit a plane in a three-dimensional space to best represent the relationships between the features (size and bedrooms) and the target (price).

y = b0 + b1 * x1 + b2 * x2

where y is the predicted price, x1 is the size of the house, x2 is the number of bedrooms, and b0, b1, and b2 are the regression coefficients.

Lasso regression and ridge regression are both regularization techniques used in linear regression to prevent overfitting by adding a penalty term to the objective function. The choice between these two regularization methods depends on the specific characteristics of your data and the model you are trying to build.

1. Lasso Regression:
Lasso (Least Absolute Shrinkage and Selection Operator) regression adds the L1 norm of the coefficients as a penalty term to the objective function. This penalty term encourages sparsity in the model, meaning it tends to drive some coefficients to exactly zero. Lasso is particularly useful when dealing with datasets that have many features, some of which may be irrelevant or redundant.

When to use Lasso:
- When you suspect that only a few features are essential for the prediction and want to perform feature selection.
- When dealing with high-dimensional data (i.e., datasets with many features).
- When you want a simple and interpretable model that only includes the most relevant features.

2. Ridge Regression:
Ridge regression adds the L2 norm of the coefficients as a penalty term to the objective function. The L2 penalty term penalizes large coefficients, effectively shrinking them towards zero without making them exactly zero. Ridge regression is helpful when multicollinearity (high correlation) exists between features, as it can help stabilize the model and reduce the impact of correlated features.

When to use Ridge:
- When you have collinear features (features that are highly correlated) and want to reduce their impact on the model.
- When you don't have prior knowledge about which features are more important, and you want to regularize all features equally.

Impact on Underfitting and Overfitting:
- Underfitting: Regularization methods like Lasso and Ridge can help reduce underfitting by preventing the model from being too simple and biased. If your model is underfitting the training data (i.e., performs poorly on both training and test data), applying either Lasso or Ridge can improve its performance by adding more flexibility without overfitting.

- Overfitting: Lasso and Ridge are both effective in preventing overfitting by penalizing large coefficients. However, Lasso's L1 penalty has the advantage of driving some coefficients to exactly zero, effectively performing feature selection, which can be more useful when dealing with high-dimensional datasets. Ridge, on the other hand, reduces the impact of correlated features but does not force any coefficients to be exactly zero.

In summary, use Lasso when you suspect that only a few features are important, or when dealing with high-dimensional data and you want feature selection. Use Ridge when you have collinear features and want to reduce their impact while still keeping all features in the model. Both methods can help prevent overfitting and underfitting, but the choice depends on the characteristics of your data and the specific goals of your modeling task. Additionally, you can also explore Elastic Net, which combines Lasso and Ridge penalties to provide a balance between feature selection and coefficient shrinkage.

>>>In summary, Lasso is used when we want to keep things simple and select only the most important features, while Ridge is used when we have similar features and want to reduce their influence. These techniques help us create models that make better predictions on new data.

Model selection: train/test split

```python
#splitting the dataset
train_set, test_set = train_test_split(mach_data, test_size=0.2)
```

Builidng a pipeline: Creating a custom tranformer to drop columns

```python
class ColumnDropper(BaseEstimator, TransformerMixin):
    def __init__(self, columns_to_drop: list):
        self.columns_to_drop = columns_to_drop

    def drop_columns(self, X):
        X = X.drop(self.columns_to_drop, axis=1)
        return(X)


    def fit(self, X:pd.DataFrame, y = None):
        return self

    def transform(self, X, y = None):
        X = self.drop_columns(X)
        return(X)
```

```python
#Executing the custom transformer
customer = ColumnDropper(["vendor name", "Model Name"])
customer.fit_transform(mach_data)
```

```python
pipe = Pipeline([("customer", ColumnDropper(["vendor name", "Model Name"]))])
pipe.fit_transform(mach_data)
```

```python
#Pipeline initialisiation
preprocess = Pipeline([
    ("DropColumns", ColumnDropper(["vendor name", "Model Name"])),
    ("Standarzation", StandardScaler()),
    ("Add-Nonelinearity", PolynomialFeatures(degree=3))
])
```

```python
#ML Models
models_pipeline = Pipeline([
    ("LinearRegs" ,LinearRegression())
])
```

```python
#Executing the pipeline
X_train = train_set[[c for c in mach_data.columns if c not in ["PRP", "ERP"]]]
Y_train = train_set[target_col]

X_test = test_set[[c for c in mach_data.columns if c not in ["PRP", "ERP"]]]
Y_test = test_set[target_col]

full_pipeline = Pipeline([
    ("Preprocess", preprocess),
    ("model", models_pipeline)
])
```

```python
 full_pipeline.fit(X_train, Y_train)
```

Model Evaluation

```python
y_hat_train = full_pipeline.predict(X_train)
y_hat_test = full_pipeline.predict(X_test)
full_pipeline.score(X_train, Y_train)
```

```python
#training set's MSE
mean_squared_error(y_hat_train, Y_train)
```

```python
#testing set's MSE
mean_squared_error(y_hat_test, Y_test)
```

## Experiment 3:
* Model Selection through bootstrapping

```python
models = []
models.append(('LR', LinearRegression()))
models.append(('LASSO', Lasso()))
models.append(('EN', ElasticNet()))
models.append(('KNN', KNeighborsRegressor()))
models.append(('CART', DecisionTreeRegressor()))
models.append(('SVR', SVR()))
```

```python
#Test options and evaluation metric
num_folds = 10
seed = 7
scoring = 'neg_mean_squared_error'
```

```python
results = []
names = []
for name, model in models:
    model_pipeline = Pipeline([
        ("Preprocess", preprocess),
        ("model", model)
    ])

    kfold = KFold(n_splits=num_folds, random_state=seed, shuffle=True)
    cv_results = cross_val_score(model_pipeline, X_train, Y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```
